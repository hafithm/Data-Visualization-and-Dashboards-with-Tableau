```python
import pandas as pd

df = pd.read_excel('/Users/hm4th/Downloads/airbnb.xlsx')

# Define the mapping of zip codes to neighborhoods
zipcode_to_neighborhood = {
    11249: 'Brooklyn',
    10024: 'Manhattan',
    11373: 'Queens',
    10312: 'Staten Island',
    10466: 'Bronx'
}

# Apply the mapping to create a new column 'Neighborhood'
df['Neighborhood'] = df['Zipcode'].map(zipcode_to_neighborhood)

# Remove all rows with any null values
df.dropna(inplace=True)

# Replace NaN with 'Other' for zip codes that weren't in the mapping
df['Neighborhood'] = df['Neighborhood'].fillna('Other')

# Now save the updated DataFrame back to an Excel file
output_file_path = '/Users/hm4th/Downloads/airbnb.xlsx'
df.to_excel(output_file_path, index=False)

```
